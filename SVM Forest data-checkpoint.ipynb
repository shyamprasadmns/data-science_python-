```python

from sklearn.model_selection import train_test_split
import sklearn.svm as svm
```

```python
X= FR.iloc[:,0:4]
y= FR.iloc[:,4]
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.3)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python

from sklearn.svm import SVC
```

```python
# kernel = rbf
model_rbf = SVC(kernel = "rbf")
model_rbf.fit(X_train,y_train)
pred_test_rbf = model_rbf.predict(X_test)

np.mean(pred_test_rbf==y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,pred_test_rbf ))
```

```python
print(classification_report(y_test,pred_test_rbf))
```

```python
# kernel = linear
model_rbf = SVC(kernel = "linear")
model_rbf.fit(X_train,y_train)
pred_test_linear = model_rbf.predict(X_test)

np.mean(pred_test_linear==y_test)
```

```python
print(confusion_matrix(y_test,pred_test_linear ))
```

```python
print(classification_report(y_test,pred_test_linear))
```

```python
# kernel = sigmoid
model_sigmoid = SVC(kernel = "sigmoid")
model_sigmoid.fit(X_train,y_train)
pred_test_sigmoid = model_sigmoid.predict(X_test)

np.mean(pred_test_sigmoid==y_test)
```

```python
print(confusion_matrix(y_test,pred_test_sigmoid ))
```

```python
print(classification_report(y_test,pred_test_sigmoid))
```

```python
model_poly = SVC(kernel = "poly",C=2)
model_poly.fit(X_train,y_train)
pred_test_poly = model_poly.predict(X_test)

np.mean(pred_test_poly==y_test)
```

```python
print(confusion_matrix(y_test,pred_test_poly ))
```
